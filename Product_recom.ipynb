```python
import pandas as pd
```

```python
df_baby=pd.read_csv('final_data.csv')
df_baby=df_baby.drop(['Unnamed: 0'],axis=1)
df_baby.head()
```

```python
import re
def cleaning(text):
    text = re.sub('[^0-9A-Za-z]+',' ',str(text))
    text=text.lower()
    text=re.sub(' +',' ',str(text))
    return text
df_baby["product_name"]=df_baby["product_name"].apply(cleaning)
```

```python
df_cloth=pd.read_csv('final_2new_combine.csv')
df_cloth=df_cloth.drop(['Unnamed: 0'],axis=1)
df_cloth.head()
```

```python
df=pd.merge(df_baby,df_cloth,on='reviewerID')
df.head(2)
```

```python
df.columns
```
